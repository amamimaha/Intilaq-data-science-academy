```python
from itertools import groupby
filename = 'How.I.Met.Your.Mother.S01E01.DVDRip.XviD-TOPAZ.EnglishHI.srt'
# "chunk" our input file, delimited by blank lines
with open(filename) as f:
    res = [list(g) for b,g in groupby(f, lambda x: bool(x.strip())) if b]
```

```python
res
```

```python
import srt
subs = []
for f in os.listdir(os.getcwd()):
    with open(filename) as f:
        data = f.read()
        subs = subs + list(srt.parse(data))
```

```python
subs
```

```python
import nltk
import re
sentences = []
regex = re.compile('[^a-zA-Z]')
for s in subs:
    sentence = regex.sub(' ', s.content)
    tokens = nltk.regexp_tokenize(sentence, "\w+")
    words = [w.lower() for w in tokens]
    #print('[%s]' % ', '.join(map(str, words)))
    sentences.append(words)
print('[%s]' % ', '.join(map(str, sentences)))
```

```python
import os
import gensim
from gensim import models
from gensim.models import Word2Vec
model = Word2Vec(sentences)
```
